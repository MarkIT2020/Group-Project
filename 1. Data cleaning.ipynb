```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set() 
df = pd.read_csv('cardio_train.csv',sep=";").drop('id', axis=1)
df.head()
df.describe()
```

```python
def calculate_BMI(row):
    new_data = row['weight']*10000/(row['height']*row['height'])
    return new_data
df['BMI'] = df.apply(calculate_BMI, axis=1)
df.to_csv('cardio.csv', index=False)

def calculate_year(row):
    new_data2 = (row['age'] / 365).round().astype('int')
    return new_data2
df['year'] = df.apply(calculate_year, axis=1)

```

```python
#remove weights and heights, that fall below 5% or above 95% of a given range
df.drop(df[(df['height'] > df['height'].quantile(0.95)) | (df['height'] < df['height'].quantile(0.05))].index,inplace=True)
df.drop(df[(df['weight'] > df['weight'].quantile(0.95)) | (df['weight'] < df['weight'].quantile(0.05))].index,inplace=True)

df.drop(df[(df['ap_hi'] > df['ap_hi'].quantile(0.95)) | (df['ap_hi'] < df['ap_hi'].quantile(0.05))].index,inplace=True)
df.drop(df[(df['ap_lo'] > df['ap_lo'].quantile(0.95)) | (df['ap_lo'] < df['ap_lo'].quantile(0.05))].index,inplace=True)
```

```python
#Create a new column 'cardio_bool' and map the values from the 'cardio' column to True or False
df['cardio_bool'] = df['cardio'].map({0: 'N', 1: 'Y'})
df.to_csv('updated_dataset.csv', index=False)
```
